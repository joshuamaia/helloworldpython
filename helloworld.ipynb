```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando um DataFrame fictício
data = {
    'Mês': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun'],
    'Vendas': [1500, 2300, 1800, 2200, 2600, 3000]
}

df = pd.DataFrame(data)

# Exibindo o DataFrame
print("Dados de vendas:")
display(df)

# Criando um gráfico de vendas
plt.figure(figsize=(8,5))
plt.plot(df['Mês'], df['Vendas'], marker='o', linestyle='-', color='b', label='Vendas')
plt.xlabel('Mês')
plt.ylabel('Vendas (R$)')
plt.title('Vendas Mensais')
plt.legend()
plt.grid()
plt.show()
```

```python
import datetime

# Definição de variáveis com tipos primitivos
inteiro = 10          # Tipo inteiro (int)
decimal = 3.14        # Tipo ponto flutuante (float)
booleano = True       # Tipo booleano (bool)
texto = "Olá, Python!" # Tipo string (str)
data_atual = datetime.datetime.now()  # Tipo datetime
data_formatada = data_atual.strftime("%d/%m/%Y %H:%M:%S")

# Imprimindo os valores e seus tipos
print(f"Valor: {inteiro}, Tipo: {type(inteiro)}")
print(f"Valor: {decimal}, Tipo: {type(decimal)}")
print(f"Valor: {booleano}, Tipo: {type(booleano)}")
print(f"Valor: '{texto}', Tipo: {type(texto)}")
print(f"Valor: {data_atual}, Tipo: {type(data_atual)}")
print(f"Valor: '{data_formatada}', Tipo: {type(data_formatada)}")
```
